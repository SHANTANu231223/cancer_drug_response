```python
import pandas as pd
from scipy.stats import chi2_contingency

# Assuming 'file' is your DataFrame
# Replace 'response_column_name' with the actual name of your dependent variable column
response_column_name = 'Rapamycin_response'
#response_column = file[response_column_name]

# Identify columns with 'mut' in their name and are binary
mut_columns = [col for col in file.columns if 'mut' in col and file[col].nunique() == 2]

# Create an empty list to store results
results_list = []

# Perform chi-square tests for each 'mut' column
for mut_column in mut_columns:
    contingency_table = pd.crosstab(file[response_column_name], file[mut_column])
    chi2, p, _, _ = chi2_contingency(contingency_table)
    
    # Append results to the list
    results_list.append({
        'Variable': mut_column,
        'Chi-Square Value': chi2,
        'P-value': p
    })

# Create a DataFrame from the list of results
results_df = pd.DataFrame(results_list)

# Display results DataFrame
print(results_df)
```

```python
results_df = results_df.rename(columns={'P-value': 'pvalue'})
```

```python
important_mut_columns=results_df[results_df.pvalue<=0.05]
important_mut_columns=important_mut_columns['Variable']
```

```python
# Calculate covariance and store results in a dictionary
covariance_dict = {}

for var in inputvar:
    covariance = np.cov(file1[outputvar], file1[var], ddof=0)[0, 1]
    covariance_dict[var] = covariance
```

```python
covariance_df = pd.DataFrame(list(covariance_dict.items()), columns=['Variable', 'Covariance'])

# Display the DataFrame
print(covariance_df)

# Save the DataFrame to a CSV file
covariance_df.to_csv('covariance_results.csv', index=False)
```

```python
# Find values less than -0.3
less_than = {key: value for key, value in covariance_dict.items() if value < -0.3}

# Find values greater than 0.3
greater_than = {key: value for key, value in covariance_dict.items() if value > 0.3}

# Display the results
print("Values less than -0.3:", less_than)
print("Values greater than 0.3:", greater_than)
```
